```python
# Training dataset
start = "2015-01-01"
end = "2022-12-31"

def get_stock_data(start, end, stock_list):
  stock_symbols = stock_list
  df = pd.DataFrame()
  for i in stock_symbols:
      stock_data = yf.Ticker(i)
      hist_stock = stock_data.history(start = start, end = end)
      eps = stock_data.get_financials().loc['BasicEPS']
      # stock selection criteria
      if eps[0] > eps.mean():
        df[i] = hist_stock["Close"]
  return df

df = get_stock_data(start, end, stock_list)
df = df.dropna(axis=1)
df
```

# Portfolio Optimizer

```python
from pypfopt.expected_returns import mean_historical_return
from pypfopt.risk_models import CovarianceShrinkage
from pypfopt.efficient_frontier import EfficientFrontier
from pypfopt import objective_functions
from pypfopt.discrete_allocation import DiscreteAllocation, get_latest_prices
import warnings

def train_optimiser(df, sector_constraints = None):
  warnings.simplefilter(action='ignore', category=UserWarning)
  mu = mean_historical_return(df)
  S = CovarianceShrinkage(df).ledoit_wolf()
  ef = EfficientFrontier(mu, S)
  ef.add_objective(objective_functions.L2_reg, gamma=0.1)
  if sector_constraints:
    ef.add_sector_constraints(**sector_constraints)
  ef.max_sharpe()
  weights = ef.clean_weights()
  warnings.resetwarnings()
  return weights
```

# Backtesting

```python
class Backtester():
  def __init__(self, benchmark, start, end, df, weights):
    self.a_results = pd.DataFrame()
    self.b_results = pd.DataFrame()
    self.benchmark_df = pd.DataFrame() # for STI stock info
    self.benchmark = benchmark # benchmark index
    self.start = start # start date
    self.end = end # end date
    self.df = df # portfolio data
    self.weights = weights # initial weights
    self.prepare_benchmark_data()

  def prepare_benchmark_data(self):
    benchmark_data = yf.Ticker(self.benchmark)
    hist_benchmark = benchmark_data.history(start = self.start, end = self.end)
    self.benchmark_df["Close_Price"] = hist_benchmark["Close"]
    self.benchmark_df["Reg_Returns"] = self.benchmark_df["Close_Price"].pct_change()
    self.benchmark_df = self.benchmark_df.dropna()

  def performance(self):

    print("="*50)
    print("Performance Metrics")
    print("-"*50)

    # stocks bought
    print("Ideal stock weightages:")
    for key, value in self.weights.items():
      if value != 0: print(f"{key}: {value}")
    print("-"*50)

    # total regular portfolio return
    self.a_results["Portfolio_Log_Returns"] = np.log(1 + self.a_results["Portfolio_Reg_Returns"])
    self.b_results["Portfolio_Log_Returns"] = np.log(1 + self.b_results["Portfolio_Reg_Returns"])
    self.benchmark_df["Portfolio_Log_Returns"] = np.log(1 + self.benchmark_df["Reg_Returns"])

    total_log_return_a = self.a_results["Portfolio_Log_Returns"].sum()
    total_log_return_b = self.b_results["Portfolio_Log_Returns"].sum()
    total_log_return_bmk = self.benchmark_df["Portfolio_Log_Returns"].sum()

    total_regular_return_a = (np.exp(total_log_return_a) - 1)*100
    total_regular_return_b = (np.exp(total_log_return_b) - 1)*100
    total_regular_return_bmk = (np.exp(total_log_return_bmk) - 1)*100

    print("Total portfolio return (a): {:0.2f} %".format(total_regular_return_a))
    print("Total portfolio return (b): {:0.2f} %".format(total_regular_return_b))
    print("Total return (benchmark): {:0.2f} %".format(total_regular_return_bmk))
    print("-"*50)

    # Annual regular returns
    annual_log_return_a = self.a_results["Portfolio_Log_Returns"].mean() * 252
    annual_log_return_b = self.b_results["Portfolio_Log_Returns"].mean() * 252
    annual_log_return_bmk = self.benchmark_df["Portfolio_Log_Returns"].mean() * 252

    annual_regular_return_a = (np.exp(annual_log_return_a) - 1)*100
    annual_regular_return_b = (np.exp(annual_log_return_b) - 1)*100
    annual_regular_return_bmk = (np.exp(annual_log_return_bmk) - 1)*100

    print("Annual regular returns (a): {:0.2f} %".format(annual_regular_return_a))
    print("Annual regular returns (b): {:0.2f} %".format(annual_regular_return_b))
    print("Annual regular returns (benchmark): {:0.2f} %".format(annual_regular_return_bmk))
    print("-"*50)

    # Annual std deviations
    daily_regular_std_a = self.a_results["Portfolio_Reg_Returns"].std()
    daily_regular_std_b = self.b_results["Portfolio_Reg_Returns"].std()
    daily_regular_std_bmk = self.benchmark_df["Reg_Returns"].std()

    annual_regular_std_a = daily_regular_std_a * (252 **0.5) *100
    annual_regular_std_b = daily_regular_std_b * (252 **0.5) *100
    annual_regular_std_bmk = daily_regular_std_bmk * (252 **0.5) *100

    print("Annual standard deviation (a): {:0.2f} %".format(annual_regular_std_a))
    print("Annual standard deviation (b): {:0.2f} %".format(annual_regular_std_b))
    print("Annual standard deviation (benchmark): {:0.2f} %".format(annual_regular_std_bmk))
    print("-"*50)

    # Sharpe ratios
    sharpe_ratio_a = annual_regular_return_a / annual_regular_std_a
    sharpe_ratio_b = annual_regular_return_b / annual_regular_std_b
    sharpe_ratio_bmk = annual_regular_return_bmk / annual_regular_std_bmk

    print("Sharpe ratio (a): {:0.2f}".format(sharpe_ratio_a))
    print("Sharpe ratio (b): {:0.2f}".format(sharpe_ratio_b))
    print("Sharpe ratio (benchmark): {:0.2f}".format(sharpe_ratio_bmk))
    print("-"*50)

    # Drawdowns
    def calculate_max_drawdown(data):
      data['Gross_Cum_Returns'] = data["Portfolio_Log_Returns"].cumsum().apply(np.exp)
      data['Cum_Max'] = data['Gross_Cum_Returns'].cummax()
      drawdown = data["Cum_Max"] - data['Gross_Cum_Returns']
      zero_periods = drawdown[drawdown == 0]
      delta_values = zero_periods.index[1:] - zero_periods.index[:-1]
      return drawdown.max()*100, delta_values.days.max()

    max_drawdown_a, max_drawdown_period_a = calculate_max_drawdown(self.a_results)
    max_drawdown_b, max_drawdown_period_b = calculate_max_drawdown(self.b_results)
    max_drawdown_bmk, max_drawdown_period_bmk = calculate_max_drawdown(self.benchmark_df)
    print("The max drawdown (a): {:0.2f} % for max {} days".format(max_drawdown_a, max_drawdown_period_a))
    print("The max drawdown (b): {:0.2f} % for max {} days".format(max_drawdown_b, max_drawdown_period_b))
    print("The max drawdown (benchmark): {:0.2f} % for max {} days".format(max_drawdown_bmk, max_drawdown_period_bmk))
    print("-"*50)

    # Calmar ratios
    calmar_ratio_a = annual_regular_return_a / max_drawdown_a
    calmar_ratio_b = annual_regular_return_b / max_drawdown_b
    calmar_ratio_bmk = annual_regular_return_bmk / max_drawdown_bmk
    print("The annual calmar ratio (a): {:0.2f}".format(calmar_ratio_a))
    print("The annual calmar ratio (b): {:0.2f}".format(calmar_ratio_b))
    print("The annual calmar ratio (benchmark): {:0.2f}".format(calmar_ratio_bmk))
    print("="*50)

  def backtest(self):
    weights = self.weights
    self.backtest_6a(weights)
    self.backtest_6b(weights)

  def backtest_6a(self, weights): # q6a
    for i in weights:
      if weights[i]:
        stock_data = yf.Ticker(i)
        hist_stock = stock_data.history(start = self.start, end = self.end)
        self.a_results[i] = hist_stock["Close"].pct_change() * weights[i]
    self.a_results['Portfolio_Reg_Returns'] = self.a_results.sum(axis=1)

  def backtest_6b(self, weights): # q6b
    latest_prices = get_latest_prices(self.df)
    da = DiscreteAllocation(weights, latest_prices, total_portfolio_value=1000000)
    allocation, leftover = da.greedy_portfolio() # Greedy
    for i in allocation:
      stock_data = yf.Ticker(i)
      hist_stock = stock_data.history(start = self.start, end = self.end)
      self.b_results[i] = hist_stock["Close"] * allocation[i]
    self.b_results['Portfolio_Value'] = self.b_results.sum(axis=1)
    self.b_results['Portfolio_Reg_Returns'] = self.b_results['Portfolio_Value'].pct_change()
```

```python
# initial run
def run_backtester(df, weights):
    bt = Backtester("^STI", "2023-01-01", "2023-11-15", df, weights)
    bt.backtest()
    bt.performance()

weights = train_optimiser(df)
run_backtester(df, weights)
```

# Sector Constraints

```python
# By Sector
sector_mapper = {}

# map tickers to sector
for stock in df:
  t = yf.Ticker(stock)
  sector_mapper[stock] = t.get_info()['sector']

print("Unique Sectors: ", set(sector_mapper.values()), '\n')

sector_lower = {
    "Consumer Cyclical": 0.60, # at least 60% to Consumer Cyclical
    "Consumer Defensive": 0.05,
    "Real Estate": 0.05,
    "Technology": 0.05,
    "Industrials": 0.05,
    "Communication Services": 0.05,
    "Financial Services": 0.05,
}
sector_upper = {
    "Consumer Defensive": 0.10, # at most 10% to Consumer Defensive
    "Real Estate": 0.10, # at most 10% to Real Estate
}

sector_constraints = {
    'sector_mapper': sector_mapper,
    'sector_lower': sector_lower,
    'sector_upper': sector_upper,
}

weights = train_optimiser(df, sector_constraints)
run_backtester(df, weights)
```
